```python
# Normalize cgpa to a scale of 10
df['normalized_cgpa'] = (df['cgpa'] / df['cgpaScale']) * 10

# Drop the original 'cgpa' and 'cgpaScale' columns if they are no longer needed
df = df.drop(columns=['cgpa', 'cgpaScale'])
```

```python
import re

separator_patterns = [',', '/',';',' or ']  
for sep in separator_patterns:
    df['specialization'] = df['specialization'].str.replace(sep, ',', regex=True)


# Explode specializations into unique rows
df['specialization'] = df['specialization'].str.strip()
df = df.assign(specialization=df['specialization'].str.split(','))
df = df.explode('specialization').reset_index(drop=True)

# Remove abnormal inputs like '???' and '\'
df['specialization'] = df['specialization'].apply(lambda x: re.sub(r'[?\\]', '', str(x)))


df['specialization'] = df['specialization'].str.lower().str.strip()


# Fill missing numerical values with their median
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())


categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    df[col] = df[col].fillna('unknown')

df = df[df['major'] != 'unknown']
df = df[df['ugCollege'] != 'unknown']


df[categorical_columns] = df[categorical_columns].apply(lambda x: x.str.lower().str.strip())
```

```python
import numpy as np

# Handling invalid and extreme values

# 1. Replace extreme values in TOEFL score (valid range: 0-120)
df['toeflScore'] = np.where((df['toeflScore'] < 0) | (df['toeflScore'] > 120), np.nan, df['toeflScore'])
df['toeflScore'].fillna(df['toeflScore'].median(), inplace=True)

# 2. Handling unrealistic CGPA values (assuming valid range: 0-10)
df['normalized_cgpa'] = np.where((df['normalized_cgpa'] < 0) | (df['normalized_cgpa'] > 10), np.nan, df['normalized_cgpa'])
df['normalized_cgpa'].fillna(df['normalized_cgpa'].median(), inplace=True)

# 3. GRE Scores: Ensure they fall within standard test limits
df['greQ'] = np.where((df['greQ'] < 130) | (df['greQ'] > 170), np.nan, df['greQ'])
df['greV'] = np.where((df['greV'] < 130) | (df['greV'] > 170), np.nan, df['greV'])
df['greA'] = np.where((df['greA'] < 0) | (df['greA'] > 6), np.nan, df['greA'])

# Fill missing GRE values with the median
df[['greQ', 'greV', 'greA']] = df[['greQ', 'greV', 'greA']].fillna(df[['greQ', 'greV', 'greA']].median())

# 4. Handling work experience values
# Assuming reasonable max values: Research <= 10, Industry <= 40, Internship <= 5
df['researchExp'] = np.where((df['researchExp'] < 0) | (df['researchExp'] > 10), np.nan, df['researchExp'])
df['industryExp'] = np.where((df['industryExp'] < 0) | (df['industryExp'] > 40), np.nan, df['industryExp'])
df['internExp'] = np.where((df['internExp'] < 0) | (df['internExp'] > 5), np.nan, df['internExp'])

# Fill missing experience values with the median
df[['researchExp', 'industryExp', 'internExp']] = df[['researchExp', 'industryExp', 'internExp']].fillna(df[['researchExp', 'industryExp', 'internExp']].median())

# Confirm data cleaning
print("\nCleaned Data Description:")
print(df[['toeflScore', 'normalized_cgpa', 'greQ', 'greV', 'greA', 'researchExp', 'industryExp', 'internExp']].describe())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("./data/categorized_specializations.csv")

df = df.drop_duplicates()

sns.set_style("whitegrid")

categorical_features = ['major', 'ugCollege', 'univ_state', 'specialization_category', 'univName']
numerical_features = ['normalized_cgpa', 'toeflScore', 'greQ', 'greV', 'greA', 'researchExp', 'industryExp', 'internExp']

# Distribution Plots for Categorical Features 
def plot_categorical_distribution(df, cat_col, title, top_n=15):
    plt.figure(figsize=(12, 5))
    ax = sns.countplot(y=df[cat_col], order=df[cat_col].value_counts().index[:top_n], palette="viridis")
    plt.title(f"Top {top_n} {title} Distribution")
    plt.xlabel("Count")
    plt.ylabel(cat_col)
    for p in ax.patches:
        ax.annotate(f'{p.get_width()}', (p.get_width(), p.get_y() + p.get_height() / 2), 
                    ha='left', va='center', fontsize=10, color='black', xytext=(5, 0), textcoords='offset points')
    plt.show()


for col in ['specialization_category', 'univName', 'major', 'ugCollege']:
    plot_categorical_distribution(df, col, col.replace("_", " ").title())
```
